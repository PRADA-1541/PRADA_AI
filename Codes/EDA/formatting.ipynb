```python
import pandas as pd
import json
import os
```

```python
# Gutekueche 리뷰데이터에 칵테일 이름을 추가하는 코드

data_path = os.path.join(os.getcwd(), "..\\..\\Dataset")
dataset_name = 'gutekueche_cocktail_reviews.json'

dataset_path = os.path.join(data_path, dataset_name)
review_df = pd.read_json(dataset_path, encoding='utf-8')
review_df
```

```python
# Gutekueche 칵테일 프로필 데이터셋 로드
dataset_name = 'gutekueche_cocktail_profiles_filtered_by_4.json'
dataset_path = os.path.join(data_path, dataset_name)

profile_df = pd.read_json(dataset_path, encoding='utf-8')
profile_df
```

```python
filtered_cocktail_list = profile_df[['ID', 'Name']]

filtered_cocktail_list

# 두 데이터셋 병합 (ID=Name 기준)
merged_df = pd.merge(left=review_df, right=filtered_cocktail_list, left_on='Cocktail', right_on='ID', how='inner')
merged_df.drop(columns=['ID'], inplace=True)
merged_df
```

```python
merged_df.info()
```

```python
# Timestamp -> Date로 변환
merged_df['Date'] = merged_df['Timestamp'].map(lambda x: x.split(" ")[1].replace(".", "-"))


merged_df = merged_df[['Cocktail', 'Name', 'User', 'Review', 'Date']]
```

```python
gutekueche_reviews = merged_df
gutekueche_reviews
```

```python
# All Recipe 리뷰 데이터 로드
dataset_name = 'all_recipe_cocktail_reviews.json'
dataset_path = os.path.join(data_path, dataset_name)

allrecipe_reviews = pd.read_json(dataset_path, encoding='utf-8')

allrecipe_reviews = allrecipe_reviews['cocktail_reviews'].to_dict()
allrecipe_reviews = pd.DataFrame(allrecipe_reviews).T

allrecipe_reviews
```

```python
# Name 대문자로 바꾸기
allrecipe_reviews['Name'] = allrecipe_reviews['Name'].map(lambda x: x.upper())

allrecipe_reviews
```

```python
# 각 데이터셋 칵테일 이름 리스트
gutekueche_cocktail_names = list(gutekueche_reviews['Name'].unique())
allrecipe_cocktail_names = list(allrecipe_reviews['Name'].unique())

# 중복되는 칵테일 이름 리스트
duplicated_cocktail_names = list(set(gutekueche_cocktail_names) & set(allrecipe_cocktail_names))
duplicated_cocktail_names
```

```python
temp_list = gutekueche_cocktail_names + allrecipe_cocktail_names
temp_list = list(set(temp_list))
print(len(temp_list))
```

```python
# IBA 칵테일 리스트 로드
dataset_name = 'IBA_cocktail_name_list.txt'
dataset_path = os.path.join(data_path, dataset_name)

iba_list = []

with open(dataset_path, 'r', encoding='utf-8') as f:
    iba_list_file = f.readlines()

for name in iba_list_file:
    name = name.strip('\n')
    name = name.upper()
    iba_list.append(name)
    
print(len(iba_list))
```

```python
# 세 데이터 셋 간 중복되는 칵테일 이름 리스트

duplicated_cocktail_names_1 = list(set(gutekueche_cocktail_names) & set(iba_list))
duplicated_cocktail_names_2 = list(set(allrecipe_cocktail_names) & set(iba_list))

print(duplicated_cocktail_names_1)
print(duplicated_cocktail_names_2)

print(len(duplicated_cocktail_names_1))
```

```python
cocktail_list = []
cocktail_list.extend(gutekueche_cocktail_names)
cocktail_list.extend(allrecipe_cocktail_names)
#cocktail_list.extend(iba_list)

cocktail_list = list(set(cocktail_list))
cocktail_list.sort()
print(len(cocktail_list))
```

```python
# 각 name의 앞글자만 대문자로 바꾸기
cocktail_list = [name.title() for name in cocktail_list]
invalid_list = [0 for _ in cocktail_list]

# Name, Alternative Name, Invalid Datframe 생성
name_df = pd.DataFrame(cocktail_list, columns=['Name'])
alternative_name_df = pd.DataFrame(cocktail_list, columns=['Alternative Name'])
invalid_df = pd.DataFrame(invalid_list, columns=['Invalid'])

# Name, Alternative Name, Invalid 데이터프레임 병합
cocktail_df = pd.concat([name_df, alternative_name_df, invalid_df], axis=1)
cocktail_df
```

```python
# cocktail_df를 json 파일로 저장
cocktail_df.to_json(os.path.join(data_path, 'cocktail_list2.json'), orient='records', force_ascii=False, indent=4)
```

```python
# cocktail_name_list.json 파일 저장
with open(os.path.join(data_path, 'cocktail_name_list.json'), 'w', encoding='utf-8') as f:
    json.dump(cocktail_list, f, ensure_ascii=False, indent=4)
```
